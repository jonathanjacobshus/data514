```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Define the nodes and their connections (edges)
G.add_edges_from([
    ("Selection (Year <= 2020) Regist", "Join Payroll and Filtered Regist on UserID"),
    ("Join Payroll and Filtered Regist on UserID", "Group By Job and Aggregation (Min Year)"),
    ("Group By Job and Aggregation (Min Year)", "Final Join: Payroll, Regist, Aggregated Results")
])

# Draw the graph
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', font_weight='bold', font_size=10, node_size=3000, edge_color='gray', arrowsize=20)
plt.title("Optimized Relational Algebra Tree")
plt.show()
```

```python
# Improved diagram with more structured layout and clearer labels

# Create a directed graph
G = nx.DiGraph()

# Nodes and labels for a clearer representation
nodes = {
    "Selection": "Selection\nσ(Year ≤ 2020) Regist",
    "Join1": "Join\nPayroll ⋈ UserID Regist",
    "GroupBy": "Group By & Aggregate\nγ(Job, Min(Year))",
    "Join2": "Final Join\nPayroll, Regist, Aggregated Results"
}

# Adding nodes with labels to the graph
for node, label in nodes.items():
    G.add_node(label)

# Define the connections (edges) between the nodes
edges = [
    (nodes["Selection"], nodes["Join1"]),
    (nodes["Join1"], nodes["GroupBy"]),
    (nodes["GroupBy"], nodes["Join2"])
]

G.add_edges_from(edges)

# Draw the graph
plt.figure(figsize=(8, 5))
pos = nx.shell_layout(G)  # shell layout for better structured layout
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=4000, font_size=9, font_weight='bold', edge_color='gray', arrows=True, arrowstyle='-|>')
plt.title("Optimized Relational Algebra Tree", size=15)
plt.show()
```
